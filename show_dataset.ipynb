## Przygotowanie danych
Przygotowujemy dane do modelowania: uzupełniamy brakujące wartości, kodujemy zmienne kategoryczne i dzielimy zbiór na część treningową oraz testową.


```python
# Check time intervals for a single patient
patient_id = df['Patient_ID'].value_counts()  # Get the most common patient ID
# patient_data = df[df['Patient_ID'] == patient_id]
# print(patient_data['Hour'].diff().dropna().unique())
patient_id
```

```python
df = df.sort_values(['Patient_ID', 'Hour']).reset_index(drop=True)
```

```python
imputation_summary = {}

# Loop through each column to check for missing values
for col in df.columns:
    if df[col].isnull().sum() > 0:
        if df[col].dtype in [np.float64, np.int64]:  # Numeric column
            if abs(df[col].skew()) > 1:  # Skewed: use median
                impute_value = df[col].median()
                df[col].fillna(impute_value, inplace=True)
                imputation_summary[col] = f'Median ({impute_value})'
            else:  # Normal-ish: use mean
                impute_value = df[col].mean()
                df[col].fillna(impute_value, inplace=True)
                imputation_summary[col] = f'Mean ({impute_value})'
        else:  # Non-numeric (categorical or object)
            impute_value = df[col].mode().iloc[0]
            df[col].fillna(impute_value, inplace=True)
            imputation_summary[col] = f'Mode ({impute_value})'

# Show summary
print("\nImputation Summary:")
for col, method in imputation_summary.items():
    print(f"{col}: {method}")

# Optional: Save the cleaned dataset
df.to_csv('Dataset_cleaned.csv', index=False)
```

```python
import numpy as np
# SBP	MAP	DBP	Resp	EtCO2
window_size = 20
features = ['HR', 'O2Sat', 'Temp', 'SBP', 'MAP', 'DBP']  # you can add others if you want
X = []
y = []

for pid, group in df.groupby('Patient_ID'):
    group = group.reset_index(drop=True)
    if len(group) < window_size:
        continue  # skip short sequences

    for i in range(len(group) - window_size + 1):
        window = group.loc[i:i+window_size-1]
        X.append(window[features].values)  # shape: (20, len(features))
        y.append(window.iloc[-1]['SepsisLabel'])  # label from the last time step
```

```python
X = np.array(X)  # shape: (num_samples, 20, num_features)
y = np.array(y)  # shape: (num_samples,)
```

```python
from sklearn.utils import resample
import numpy as np

# Convert to NumPy arrays if not already
X = np.array(X)
y = np.array(y)

# Separate majority and minority classes
X_pos = X[y == 1]
X_neg = X[y == 0]

# Downsample majority class
n_pos = len(X_pos)
X_neg_downsampled = resample(X_neg, replace=False, n_samples=n_pos, random_state=42)

# Combine balanced data
X_balanced = np.concatenate((X_pos, X_neg_downsampled), axis=0)
y_balanced = np.concatenate((np.ones(n_pos), np.zeros(n_pos)), axis=0)

# Shuffle the balanced data
from sklearn.utils import shuffle
X_balanced, y_balanced = shuffle(X_balanced, y_balanced, random_state=42)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_val, y_train, y_val = train_test_split(
    X_balanced, y_balanced, test_size=0.2, stratify=y_balanced, random_state=42)
```

```python
import keras_tuner as kt
from tensorflow.keras import Input
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense

def build_model(hp):
    model = Sequential()
    model.add(Input(shape=(20, len(features))))

    # Tune LSTM units
    model.add(LSTM(units=hp.Int('lstm_units', min_value=32, max_value=128, step=32), return_sequences=False))

    # Optional Dense Layer
    if hp.Boolean("add_dense"):
        model.add(Dense(units=hp.Int('dense_units', 16, 64, step=16), activation='relu'))

    # Output layer
    model.add(Dense(1, activation='sigmoid'))

    # Tune learning rate
    model.compile(
        optimizer=tf.keras.optimizers.Adam(
            learning_rate=hp.Float("lr", min_value=1e-4, max_value=1e-2, sampling="log")),
        loss='binary_crossentropy',
        metrics=['accuracy', tf.keras.metrics.AUC(name='auc')]
    )

    return model
```

```python
import tensorflow as tf
from tensorflow.keras.callbacks import EarlyStopping

tuner = kt.BayesianOptimization(
    build_model,
    objective='val_auc',
    max_trials=15,
    executions_per_trial=1,
    directory='bayes_tuning',
    project_name='sepsis_lstm'
)

early_stop = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)

tuner.search(
    X_train, y_train,
    validation_data=(X_val, y_val),
    epochs=30,
    callbacks=[early_stop],
    verbose=1
)

best_model = tuner.get_best_models(num_models=1)[0]
best_hp = tuner.get_best_hyperparameters(num_trials=1)[0]

print("Best hyperparameters:")
for param in best_hp.values:
    print(f"{param}: {best_hp[param]}")

```

# Zbalansowanie danych wejściowych - SMOTE
Największym problemem powyzszych modeli, było zdecydowane niezbilansowanie danych wejściowych. Ponizej stosujemy SMOTE, aby zbilansować te dane. 

```python
from imblearn.over_sampling import SMOTE
from collections import Counter

# Flatten the time-series window for SMOTE
num_samples, window_size, num_features = X.shape
X_2d = X.reshape((num_samples, window_size * num_features))

# Apply SMOTE
smote = SMOTE(random_state=42)
X_resampled_2d, y_resampled = smote.fit_resample(X_2d, y)

# Reshape back to 3D
X_resampled = X_resampled_2d.reshape((-1, window_size, num_features))

# Check class distribution
print("Original class distribution:", Counter(y))
print("Resampled class distribution:", Counter(y_resampled))
```

### Ostateczny model

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix

# Split the data
X_train, X_test, y_train, y_test = train_test_split(
    X_resampled, y_resampled, test_size=0.2, random_state=42, stratify=y_resampled
)

# Define the model
model = Sequential([
    LSTM(64, input_shape=(window_size, num_features), return_sequences=True),
    Dropout(0.3),
    LSTM(32),
    Dropout(0.3),
    Dense(16, activation='relu'),
    Dense(1, activation='sigmoid')
])

# Compile the model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Train the model
history = model.fit(
    X_train, y_train,
    validation_data=(X_test, y_test),
    epochs=10,
    batch_size=64,
    verbose=1
)
```

### Ocena ostatecznego modelu

```python
# Predict
y_pred_probs = model.predict(X_test)
y_pred = (y_pred_probs > 0.5).astype(int) 

# Metrics
print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
import matplotlib.pyplot as plt

plt.plot(history.history['accuracy'], label='Train Acc')
plt.plot(history.history['val_accuracy'], label='Val Acc')
plt.title('Model Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```

### Testowanie skuteczności róznorodnych thresholdów

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import precision_recall_fscore_support, f1_score

# Predict probabilities (already done)
# y_pred_probs = model.predict(X_test)

# Ground truth labels
# y_test = ...

# Threshold range
thresholds = np.arange(0.0, 1.01, 0.01)

# Store metrics
precisions = []
recalls = []
f1s = []

for thresh in thresholds:
    y_pred_thresh = (y_pred_probs > thresh).astype(int)
    precision, recall, f1, _ = precision_recall_fscore_support(y_test, y_pred_thresh, average='binary', zero_division=0)
    precisions.append(precision)
    recalls.append(recall)
    f1s.append(f1)

# Plotting
plt.figure(figsize=(10,6))
plt.plot(thresholds, precisions, label='Precision')
plt.plot(thresholds, recalls, label='Recall')
plt.plot(thresholds, f1s, label='F1 Score')
plt.xlabel('Threshold')
plt.ylabel('Score')
plt.title('Precision, Recall, and F1 Score vs. Threshold')
plt.legend()
plt.grid()
plt.show()

# Find threshold with best F1
best_index = np.argmax(f1s)
best_thresh = thresholds[best_index]
print(f"Best Threshold: {best_thresh:.2f} — F1 Score: {f1s[best_index]:.4f}")
```
